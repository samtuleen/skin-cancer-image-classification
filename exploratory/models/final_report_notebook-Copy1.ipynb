### Model Visualization Function:

Now I'll define a function that plots the accuracy and loss results of my CNN models for easy interpretation. The results will display the accuracy and the loss in one visual:

```python
def model_performance(history):
    
    history_dict = history.history
    train_loss_values = history_dict['loss']
    val_loss_values = history_dict['val_loss']#<--'val' here is for validation, and will be from the 'test' dataset
    train_accuracy = history_dict['accuracy']
    val_accuracy = history_dict['val_accuracy']

    fig, axis = plt.subplots(ncols=1, nrows=2, figsize=(8,8))
    
    #Plotting loss results:    
    epochs = range(1, len(val_loss_values) + 1) #epoch: the number of times the model will be fit on the training set.
    plot1 = sns.lineplot(ax=axis[0], x=epochs, y=train_loss_values, label='Train Loss')
    plot1 = sns.lineplot(ax=axis[0], x=epochs, y=val_loss_values, label='Val Loss')
    plot1.set(xlabel='Epochs', ylabel='Loss')
    plot1.axes.set_title('Model Loss', fontsize=18)
    plot1.grid(b=True, which='major', axis='y') #To show the grid lines or not
    
    plot2 = sns.lineplot(ax=axis[1], x=epochs, y=train_accuracy, label='Train Accuracy')
    plot2 = sns.lineplot(ax=axis[1], x=epochs, y=val_accuracy, label='Val Accuracy')
    plot2.set(xlabel='Epochs', ylabel='Accuracy')
    plot2.axes.set_title('Model Accuracy', fontsize=20)
    plot2.grid(b=True, which='major', axis='y')
    
    plt.tight_layout()
    plt.show()
```

# Five-Block CNN Models: VGG16 Architecture

To start off this project I will begin by building 2 models using the VGG16 architecture of 5 convolutional blocks. To use these models, I will augment the data even though my dataset is decently sized (12,500 images) because CNNs perform best with more data.

## Image Augmentation:

Because I've gathered the data from numerous sources, I've examined the sizes and noticed that they all vary with an overwhelming majority of them  being as large as 350 pixels wide and 400 pixels high (with a few exceptions). I've also observed that most of the images are of the lesion only (minimal background space) and are not all centered, so I've decided against cropping and have chosen to squeeze (resize) to the standard 224x224 size for a uniform size to feed the model.

I will rescale the data  to multiply the it before any other processing. My original images consist of RGB coefficients in the 0-255 range, but these values are too high for my models to process (given a typical learning rate), so I'll target values between 0 and 1 instead by scaling with a 1/255. factor. (2)
 Additionally I will apply a zoom_range of 0.3 to randomly zoom as well as and vertically flip  some of the images on random. I've made these choices because they are relevant when there are no assumptions of vertical assymetry or a set zoom (e.g. real-world pictures are taken at different angles and distances and our model should be able to handle them all).
 
 Since skin lesions come in many different shapes and sizes, by vertically flipping and zooming we will still produce lesions that we would expect our model to classify correctly. These two transformations will help our model greatlyby inflating our dataset, increase model robustness, and help prevent overfitting.
 
 After testing various batch sizes, I've settled on a batch_size of 20 since it offered the best results and the most efficiency.
 
 And lastly, I will be using the .flow_from_directory() method that will take a path of a directory and generate our augmented data. (3)
 
 I will define a function "data_processing" to perform the augmentation below:

```python
def data_processing(img_dims, batch_size):
    train_data_gen = ImageDataGenerator(rescale=1./255, zoom_range=0.3, vertical_flip=True)
    test_data_gen = ImageDataGenerator(rescale=1./255, zoom_range=0.3, vertical_flip=True) #scaling the images 
    #to the same range [0,1]
    
    # This is fed to the network in the specified batch sizes and image dimensions
    train_generator = train_data_gen.flow_from_directory(#takes path of a directory and generates batches of augmented data
    directory=train_dir, 
    target_size=(img_dims, img_dims),  #Each image will be resized to this size
    batch_size=batch_size, #number of observations that are inputted into the model at one time
    class_mode='binary', #since there are only 2 classes, this is a binary problem
    shuffle=True)

    test_generator = test_data_gen.flow_from_directory(
    directory=test_dir, 
    target_size=(img_dims, img_dims), 
    batch_size=batch_size, 
    class_mode='binary', 
    shuffle=False)
    
    # In order to plot the Confusion Matrix, I will be making predictions off of the test set in one batch size
    test_data = []
    test_labels = []

    for cond in ['/benign/', '/malignant/']:
        for img in (os.listdir(main_dir + 'test' + cond)):
            img = plt.imread(main_dir+'test'+cond+img)
            img = cv2.resize(img, (img_dims, img_dims))
            img = np.dstack([img, img, img])
            img = img.astype('float32') / 255
            if cond=='/benign/': #setting labels for our classes: 0 for benign, 1 for malignant
                label = 0
            elif cond=='/malignant/':
                label = 1
            test_data.append(img)
            test_labels.append(label)
        
    test_data = np.array(test_data) #converting to a numpy array for quick calculations
    test_labels = np.array(test_labels)
    
    return train_generator, test_generator, test_data, test_labels
```

```python
#Values for augmentation and later use:
img_dims = 224
epochs = 30
batch_size = 20
train_generator, test_generator, test_data, test_labels = data_processing(img_dims, batch_size)
```

## Model 1:

#### Defining the model architecture and compiling (4):

I will build my first model using the following 5 steps:

1- Use five convolutional blocks comprised of convolutional layers, BatchNormalization, and MaxPooling.

2- Add a flattened layer followed by fully connected layers.

3- Reduce over-fitting by using dropouts.

4- Use ReLu (rectified linear unit) activation function for all except the last layer. Since this is a binary classification problem, Sigmoid was used for the final layer.

5- Use the Adam optimizer and cross-entropy for the loss.


 For the first block I will use a standard Conv2d (Convolutional 2D: a spatial convolution over images) layer that creates a convolution kernel that is convolved with the layer input to produce a tensor of outputs. (5) This layer takes in the input shape when used as the first layer of the model. 

I will use ReLu (Rectified Linear Unit) for my activation layer which is a piecewise linear function that outputs zero if its input is negative, and directly outputs the input otherwise.(6)

I set the padding to 'same'. Padding is to add extra pixels around the image since every we use the kernel the image will get smaller. By setting it to "same", I am adding layers to keep the output image's dimensions the same as the inputs (7)
 

I add a MaxPool layer to reduce the dimensionality of images by reducing the number of pixels in the output from the previous convolutional layer by setting the pool size to 2, 2. (8)

For the second and subsequent convolutions I will use a combination of SeparableConv2D layers (see below definition) and BatchNormalization.

And for the final output layer I will use 'sigmoid' for the activation since its gradient is defined everywhere and its output is conveniently between 0 and 1 for all x. This means it squashes some input (generally the z value in a NN) between 0 and 1, where large positive values converge to 1, and large negative values converge to 0. (9)

**Spatial Convolution Definition:** Convolution is an operation on two functions f and g, which produces a third function that can be interpreted as a modified ("filtered") version of f. In this interpretation we call g the filter. If f is defined on a spatial variable like x rather than a time variable like t, we call the operation spatial convolution. Convolution lies at the heart of any physical device or computational procedure that performs smoothing or sharpening. Applied to two dimensional functions like images, it's also useful for edge finding, feature detection, motion detection, image matching, and countless other tasks. (10)

**Separable Convolutional 2D definition:**  deals not just with the spatial dimensions, but with the depth dimension  the number of channels  as well. A depthwise separable convolution splits a kernel into 2 separate kernels that do two convolutions: the depthwise convolution and the pointwise convolution. SeparableConv2D layers are intended for faster computation. (11)

**BatchNormalization definintion:** a method used to make artificial neural networks faster and more stable through normalization of the layers' inputs by re-centering and re-scaling. 

```python
inputs = Input(shape=(img_dims, img_dims, 3))

#1st Convolution
model = Conv2D(filters=16, kernel_size=(3, 3), activation='relu', padding='same')(inputs)
model = Conv2D(filters=16, kernel_size=(3, 3), activation='relu', padding='same')(model)
model = MaxPool2D(pool_size=(2, 2))(model)

#2nd Convolution
model = SeparableConv2D(filters=32, kernel_size=(3, 3), activation='relu', padding='same')(model) #(11, 11A)
model = SeparableConv2D(filters=32, kernel_size=(3, 3), activation='relu', padding='same')(model)
model = BatchNormalization()(model)
model = MaxPool2D(pool_size=(2, 2))(model)

#3rd Convolution
model = SeparableConv2D(filters=64, kernel_size=(3, 3), activation='relu', padding='same')(model)
model = SeparableConv2D(filters=64, kernel_size=(3, 3), activation='relu', padding='same')(model)
model = BatchNormalization()(model)
model = MaxPool2D(pool_size=(2, 2))(model)

#4th Convolution
model = SeparableConv2D(filters=128, kernel_size=(3, 3), activation='relu', padding='same')(model)
model = SeparableConv2D(filters=128, kernel_size=(3, 3), activation='relu', padding='same')(model)
model = BatchNormalization()(model)
model = MaxPool2D(pool_size=(2, 2))(model)
model = Dropout(rate=0.2)(model)

#5th Convolution
model = SeparableConv2D(filters=256, kernel_size=(3, 3), activation='relu', padding='same')(model)
model = SeparableConv2D(filters=256, kernel_size=(3, 3), activation='relu', padding='same')(model)
model = BatchNormalization()(model)
model = MaxPool2D(pool_size=(2, 2))(model)
model = Dropout(rate=0.2)(model)

# Fully Connected layer
model = Flatten()(model)
model = Dense(units=512, activation='relu')(model)
model = Dropout(rate=0.6)(model)
model = Dense(units=128, activation='relu')(model)
model = Dropout(rate=0.4)(model)
model = Dense(units=64, activation='relu')(model)
model = Dropout(rate=0.2)(model)

# Output layer
output = Dense(units=1, activation='sigmoid')(model) #<--Converting a real value into a interpretable probability

# Creating model and compiling
model = Model(inputs=inputs, outputs=output)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Callbacks
checkpoint = ModelCheckpoint(filepath='best_weights.hdf5', save_best_only=True, save_weights_only=True)
#hdf5 (Hierarchical Data Format) is to store the weights of the best performing epoch
lr_reduce = ReduceLROnPlateau(monitor='val_loss', factor=0.3, patience=2, verbose=2, mode='max') #<--Reduce learning rate 
#when the metric has stopped improving.
early_stop = EarlyStopping(monitor='val_loss', min_delta=0.1, patience=1, mode='min')
```

And now I'll fit the model using the callbacks I've defined above: saving the best weights as a checkpoint and reducing the learning rate when the validation loss worsens:

```python
hist = model.fit(train_generator, steps_per_epoch=train_generator.samples // batch_size, 
           epochs=epochs, validation_data=test_generator, shuffle=True,
           validation_steps=test_generator.samples // batch_size, callbacks=[checkpoint, lr_reduce])
```

Using the function defined above I'll check the perfomance:

```python
model_performance(hist)
```

Now I'll calculate the confusion matrix:

```python
preds = model.predict(test_generator)

accur = accuracy_score(test_labels, np.round(preds))*100
cm = confusion_matrix(test_labels, np.round(preds))
tn, fp, fn, tp = cm.ravel()
#tp = true positive tn = true negative fp = false positive fn = false negative


print('CONFUSION MATRIX ------------------')
print(cm)

print('\nTEST METRICS ----------------------')
precision = tp/(tp+fp)*100
recall = tp/(tp+fn)*100
print('Accuracy: {}%'.format(accur))
print('Precision: {}%'.format(precision))
print('Recall: {}%'.format(recall))
print('F1-score: {}'.format(2*precision*recall/(precision+recall)))

print('\nTRAIN METRIC ----------------------')
print('Train Accuracy: {}'.format(np.round((hist.history["accuracy"][-1])*100, 2)))
```

And finally visualize the matrix:

```python
cm = pd.DataFrame(cm, index = ['Actual Benign', 'Actual Malignant'], columns = ['Predicted Benign', 'Predicted Malignant'])
plt.figure(figsize = (8,7))
labels = ['Benign', 'Malignant']
sns.heatmap(cm, cmap= "Blues", linecolor = 'black' , linewidth = 1, annot = True, fmt='')
plt.xlabel('Predicted', fontsize=14)
plt.ylabel('True', fontsize=14)
plt.show()
```

Now on to defining my model's architecture.

For my final activation layer I'll use 'softmax'. 'Softmax' normalizes outputs by assigning probabilities to the sum of weighted values such that the sum of probabilities is equal to 1, so that each class has its own probability distribution in the class membership.():

```python
model3 = Sequential()

model3.add(Conv2D(32, (3,3), input_shape = (input_shape), activation = 'relu'))
model3.add(MaxPooling2D(pool_size = (2, 2)))
model3.add(Dropout(0.2))


model3.add(Conv2D(32, (3,3), activation = 'relu'))
model3.add(MaxPooling2D(pool_size = (2, 2)))
model3.add(Dropout(0.2))

model3.add(Conv2D(64, (3,3), activation = 'relu'))
model3.add(MaxPooling2D(pool_size = (2, 2)))
model3.add(Dropout(0.2))

model3.add(Flatten())

model3.add(Dense(units = 4096, activation = 'relu'))
model3.add(Dense(units = 2048, activation = 'relu'))
model3.add(Dropout(0.2))

model3.add(Dense(units = len(label), activation = 'softmax'))


model3.summary()
```

```python
#Resizing the images to a uniform size (a bit smaller than the usual 224 to save some memory space)
def resize_images(directory, resize):
    images = []
    read = lambda imname: np.asarray(Image.open(imname).convert("RGB"))
    for image_name in os.listdir(directory):
        PATH = os.path.join(directory, image_name)
        _, ftype = os.path.splitext(PATH)
        if ftype.endswith((".jpg", ".jpeg", ".png")):
            img = read(PATH)
            img = cv2.resize(img, (resize, resize))
            images.append(np.array(img)/255.) #<-- EfficientNet models require inputs to be float tensors of pixels
            #w/ values in the [0-255] range.
    return images
```

```python
base_model= ResNet50(include_top=False, weights="imagenet", input_shape=(height, width,3))
model5= Sequential()
model5.add(base_model)
model5.add(Conv2D(64, (3, 3), activation = 'relu'))
model5.add(Dropout(0.2))
model5.add(Conv2D(64, (3, 3), activation = 'relu'))
model5.add(MaxPooling2D(pool_size = (2, 2)))
model5.add(Dropout(0.2))
model5.add(Flatten())
model5.add(Dense(512,activation='relu'))
model5.add(Dense(256,activation='relu'))
model5.add(Dropout(0.2))
model5.add(Dense(128,activation='relu'))
model5.add(Dense(64,activation='relu'))
model5.add(Dense(32,activation='relu'))
model5.add(Dense(16,activation='relu'))
model5.add(Dense(8,activation='relu'))
model5.add(Dense(2, activation='softmax'))
model5.summary()
```
